```python
import pandas as pd
```

```python
ratings_movies = pd.read_csv('data/ratings_movies.csv', sep = ',')
ratings_movies = ratings_movies.drop('Unnamed: 0', axis=1)
ratings_movies.value_counts().shape[0]
```

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python

ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
ratings_movies.info()
ratings_movies.tail(6)
```

Orders and Products

```python
orders = pd.read_csv('data/orders.csv', sep=';')
products = pd.read_csv('data/products.csv', sep=';')
```

```python
orders
```

```python
products.head()
```

```python
orders_products = orders.merge(products,left_on='ID товара', right_on='Product_ID',how='left')
orders_products
```

```python
orders_products['Сумма заказа'] = orders_products['Количество']*orders_products['Price']

orders_products.groupby('ID Покупателя')['Сумма заказа'].sum().sort_values(ascending=False)

```
